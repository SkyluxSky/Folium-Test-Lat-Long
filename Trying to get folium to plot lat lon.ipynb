```python
pd.set_option('display.max_rows', 500)
df = pd.read_csv('Value_of_Energy_Cost_Savings_Program_Savings_for_Businesses.csv')

df
```

```python
newyork = df.sort_values(by = 'City')

newyork = newyork[(newyork['City'] == 'New York')]
```

```python
newyork = newyork[newyork['Latitude'].notnull() & newyork['Longitude'].notnull()]
list_of_coordinates = newyork.values.tolist()
print(list_of_coordinates)
newyork
```

```python
new = newyork[(newyork['City'] == 'New York')]

m = folium.Map(location=[40.738, -73.98],
                        zoom_start=13)

for index, row in newyork.iterrows():
    folium.CircleMarker([float(row['Latitude']), float(row['Longitude'])],
                        radius=15,
                        popup=row['Company Name'],
                        fill_color="#3db7e4", # divvy color
                       ).add_to(m)
    
m    
```
